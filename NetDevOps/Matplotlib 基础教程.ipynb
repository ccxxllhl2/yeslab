```python
from matplotlib import pyplot as plt
import numpy as np
```

## 创建数据

```python
A = np.array([[1,2],[3,5],[2,4]])
B = np.array([[5,2],[1,6],[3,9],[7,3]])
```

## 建立绘图
* 坐标系制作
* 数据分布
* 函数编写
* 画散点图
* 画函数图

```python
def plot_point(a_array, b_array):
    # 坐标系制作
    plt.xlim((0,10))
    plt.ylim((0,10))
    plt.xlabel('x')
    plt.ylabel('y')
    
    # 数据分布
    X = np.linspace(0,10,100)
    
    # 函数编写
    func_y = 3*X+1
    
    # 画散点图
    plt.plot(a_array[:,0], a_array[:,1], 'ro')
    plt.plot(b_array[:,0], b_array[:,1], 'gx')
    
    # 函数图
    plt.plot(X, func_y, 'b-')
    
    # 合理显示图像
    plt.show()
```

```python
plot_point(A,B)
```
